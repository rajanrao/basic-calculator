```python
from tkinter import * 
root=Tk() #creating of a window
root.title('shubham\'s and son Co.\xa9')

def calculate(number):
    global operator
    operator=operator+str(number)
    text_input.set(operator)
    
def clear():
    global operator
    text_input.set('ek baar aur shubham')
    operator=''
    
def equal():
    global operator
    sumup=float(eval(operator))
    text_input.set(sumup)
    operator=''



text_input=StringVar(value='shubhamji lets start')
operator=''
display=Entry(root,font=('arial',30,'bold'),bg='yellow',fg='black',bd=10,justify='right',text=text_input)
display.grid(columnspan=4)


##############row1##############

btn7=Button(root,padx=40,pady=30,font=('arial',30,'bold'),text='7',bg='white',fg='black',bd=10,command=lambda:calculate(7)).grid(row=1,column=0)
btn8=Button(root,padx=40,pady=30,font=('arial',30,'bold'),text='8',bg='white',fg='black',bd=10,command=lambda:calculate(8)).grid(row=1,column=1)
btn9=Button(root,padx=40,pady=30,font=('arial',30,'bold'),text='9',bg='white',fg='black',bd=10,command=lambda:calculate(9)).grid(row=1,column=2)
btnclear=Button(root,padx=40,pady=30,font=('arial',30,'bold'),text='C',bg='red',fg='black',bd=10,command=clear).grid(row=1,column=3)

###############row2###################
btn4=Button(root,padx=40,pady=30,font=('arial',30,'bold'),text='4',bg='white',fg='black',bd=10,command=lambda:calculate(4)).grid(row=2,column=0)
btn5=Button(root,padx=40,pady=30,font=('arial',30,'bold'),text='5',bg='white',fg='black',bd=10,command=lambda:calculate(5)).grid(row=2,column=1)
btn6=Button(root,padx=40,pady=30,font=('arial',30,'bold'),text='6',bg='white',fg='black',bd=10,command=lambda:calculate(6)).grid(row=2,column=2)
btndiv=Button(root,padx=40,pady=30,font=('arial',30,'bold'),text='/',bg='green',fg='black',bd=10,command=lambda:calculate('/')).grid(row=2,column=3)

################row3##################
btn1=Button(root,padx=40,pady=30,font=('arial',30,'bold'),text='1',bg='white',fg='black',bd=10,command=lambda:calculate(1)).grid(row=3,column=0)
btn2=Button(root,padx=40,pady=30,font=('arial',30,'bold'),text='2',bg='white',fg='black',bd=10,command=lambda:calculate(2)).grid(row=3,column=1)
btn3=Button(root,padx=40,pady=30,font=('arial',30,'bold'),text='3',bg='white',fg='black',bd=10,command=lambda:calculate(3)).grid(row=3,column=2)
btnminus=Button(root,padx=40,pady=30,font=('arial',30,'bold'),text='-',bg='green',fg='black',bd=10,command=lambda:calculate('-')).grid(row=3,column=3)

##############row4###################
btn0=Button(root,padx=40,pady=30,font=('arial',30,'bold'),text='0',bg='white',fg='black',bd=10,command=lambda:calculate(0)).grid(row=4,column=0)
btnmul=Button(root,padx=40,pady=30,font=('arial',30,'bold'),text='*',bg='green',fg='black',bd=10,command=lambda:calculate('*')).grid(row=4,column=1)
btnplus=Button(root,padx=40,pady=30,font=('arial',30,'bold'),text='+',bg='green',fg='black',bd=10,command=lambda:calculate('+')).grid(row=4,column=2)
btndot=Button(root,padx=45,pady=35,font=('arial',20,'bold'),text='.',bg='green',fg='black',bd=10,command=lambda:calculate('.')).grid(row=4,column=3)
################row5###############

btnopen=Button(root,padx=40,pady=30,font=('arial',30,'bold'),text='(',bg='green',fg='black',bd=10,command=lambda:calculate('(')).grid(row=5,column=0)
btnclose=Button(root,padx=40,pady=30,font=('arial',30,'bold'),text=')',bg='green',fg='black',bd=10,command=lambda:calculate(')')).grid(row=5,column=1)
btnequal=Button(root,padx=100,pady=30,font=('arial',30,'bold'),text='=',bg='green',fg='black',bd=10,command=equal).grid(row=5,column=2,columnspan=2)
root.mainloop() #end of the execution

```
